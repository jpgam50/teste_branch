```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats
from matplotlib.gridspec import GridSpec

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import QuantileTransformer
from sklearn.preprocessing import PowerTransformer
```

```python
#leitura dos dados de treinamento
df = pd.read_csv("dados_treinamento.csv" ,sep=',');
print("Formato tabela: ",df.shape)
df.head(3)
```

## Variáveis numéricas

```python
#Dúvidas:
# -a variável 'age' possui 44 valores diferentes. O que seria melhor
#neste caso, considerá-la contínua ou discreta?

# -qual seria a consequência de considerá-la discreta?
```

```python
#Seleciona variáveis numéricas, discretas e contínuas
vars_num = [var for var in df.columns if df[var].dtype != 'O']
vars_discr=[var for var in vars_num if len(df[var].unique()) < 40]
vars_count=[var for var in vars_num if var not in vars_discr]

print("Numéricas: \n",vars_num)
print("\nDiscretas: \n",vars_discr)
print("\nContínuas: \n",vars_count)
```

```python
# Histogramas (distribuições):
df[vars_count].hist(bins=60, figsize=(15,15))
plt.show()
```

## Aplicas transformações

```python
    
def escreve_corr(ax,x,y):
    x[x>0]
    y[y>0]
    cor_posison=stats.pearsonr(x, y)[0]
    xmin, xmax = ax.get_xlim()    
    ymin, ymax = ax.get_ylim()    
    dx= 0.1*(xmax-xmin)              #calcula 10% do intervalo
    dy= 0.1*(ymax-ymin)              #calcula 10% do intervalo  
    print('ymax: ',ymax)
    print(xmin+dx)
    print(ymax+dy)
    ax.text(xmin+dx, ymax-3*dy, "{:.2f}".format(cor_posison) , fontsize=15, ha='center', color="navy")
    ax.text(xmin+dx, ymax-3*dy, "Correlação de pearson", fontsize=13, ha='center')
    
```

```python
#livro: estatistica multivariada
#https://www.amazon.com.br/An%C3%A1lise-Estat%C3%ADstica-Multivariada-Abordagem-Aplicada/dp/857041451X/ref=pd_sbs_sccl_2_2/145-3082532-9429566?pd_rd_w=5fFh3&content-id=amzn1.sym.d27ac626-cb3c-4c30-830f-0fd405f2510f&pf_rd_p=d27ac626-cb3c-4c30-830f-0fd405f2510f&pf_rd_r=BCVXX0VGHNV82QC072K4&pd_rd_wg=jitH6&pd_rd_r=f37ffab6-8bdf-436a-be88-93b82ef0a008&pd_rd_i=857041451X&psc=1

listaVar=['age',
 'height_cm',
 'weight_kg',
 'body fat_%',
 'diastolic',
 'systolic',
 'gripForce',
 'sit and bend forward_cm',
 'sit-ups counts',
 'broad jump_cm']

for var in listaVar:
    
    fig = plt.figure(constrained_layout=True, figsize=(14, 7))  #cria a figura
    gs = GridSpec(2, 3, figure=fig)
    ax1 = fig.add_subplot(gs[0, 0])
    ax2 = fig.add_subplot(gs[0, 1])
    ax3 = fig.add_subplot(gs[0, 2])
    ax4 = fig.add_subplot(gs[1, 0])  
    ax5 = fig.add_subplot(gs[1, 1])
    ax6 = fig.add_subplot(gs[1, 2])   

    #variável alvo
    target=df['class']          
        
    #variável X sem transformações
    X=df[var].values.reshape(-1, 1)
    plt.title(var)
    #variável X com transformações (standart, quantile, yeo)
    X_standard= StandardScaler().fit_transform(X)
    X_quantile= QuantileTransformer(output_distribution="normal").fit_transform(X)
    X_yeo= PowerTransformer(method="yeo-johnson").fit_transform(X)

    #plota histogramas
    ax1.hist(X, bins=50)          #sem transformaçõa
    ax2.hist(X_quantile, bins=50) #com transf. quantile
    ax3.hist(X_yeo, bins=50)      #com transf. yeo

    #plota correlações 
    l1=pd.DataFrame(X,columns=['X'])
    l1['class']=target    
    sns.boxplot(x='class', y='X', data=l1,
                palette="Blues", ax=ax4,
                linewidth=1.2, fliersize=2, width=0.5)     
    
    l1=pd.DataFrame(X_quantile, columns=['X_quantile'])
    l1['class']=target    
    sns.boxplot(x='class', y='X_quantile', data= l1,
                palette="Blues", ax=ax5,
                linewidth=1.2, fliersize=2, width=0.5)   
    
    l1=pd.DataFrame(X_yeo,columns=['X_yeo'])
    l1['class']=target        
    sns.boxplot(x='class', y='X_yeo', data=l1,
                palette="Blues", ax=ax6,
                linewidth=1.2, fliersize=2, width=0.5)  

    plt.show()
```
